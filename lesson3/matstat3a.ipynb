```python
#  Домашнее задание к уроку 3

import numpy as np
from math import factorial

def sum_numbers(numbers):
    summa = 0
    for number in numbers:
        summa += number
    return summa

def sum_numbers_square(numbers, num_aver):
    summa = 0
    for number in numbers:
        summa += (number-num_aver)**2
    return summa
def comb(n, k):
    return int(factorial(n)/(factorial(k)*factorial(n-k)))


#Задача 1

z = np.array([100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150])
x = [100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]

n = len(x)
avr = sum_numbers(x) / n
print(f'Мат ожидание расчетное {avr}, по numpy {np.mean(z)}')

d = sum_numbers_square(x,avr)

d1 = d / n
d2 = d / (n - 1)

print(f'Смещенная оценка дисперсии расчетная {d1}, по numpy {np.var(z)}')
print(f'Несмещенная оценка дисперсии расчетная {d2},по numpy {np.var(z, ddof=1)}')

print(f'Смещенная оценка СКО расчетная {d1**0.5}, по numpy {np.std(z)}')
print(f'Несмещенная оценка СКО расчетная {d2**0.5},по numpy {np.std(z, ddof=1)}')


```
